```python
class ATM:
    def __init__(self):
        self.balance = 10000  # Initial balance

    def check_balance(self):
        return f"Your balance is Rs. {self.balance}"

    def withdraw(self, amount):
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
            return f"Withdrawal successful. Remaining balance: Rs. {self.balance}"
        else:
            return "Insufficient balance or invalid amount"

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposit successful. New balance: Rs. {self.balance}"
        else:
            return "Invalid amount"

def main():
    atm = ATM()
    pin = input("Enter your PIN: ")

    if pin == "1234":  # Sample PIN
        while True:
            print("\nOptions you can Exercise are:")
            print("1) Balance")
            print("2) Withdraw")
            print("3) Deposit")
            print("4) Exit")

            choice = input("Select Your Transaction from the above options: ")

            if choice == '1':
                print(atm.check_balance())
            elif choice == '2':
                amount = float(input("Enter Amount: "))
                print(atm.withdraw(amount))
            elif choice == '3':
                amount = float(input("Enter Amount: "))
                print(atm.deposit(amount))
            elif choice == '4':
                print("Thank you for using the ATM. Goodbye!")
                break
            else:
                print("Invalid choice")
    else:
        print("Invalid PIN")

if __name__ == "__main__":
    print("***** Welcome To ATM Machine Simulator *****")
    main()
```
